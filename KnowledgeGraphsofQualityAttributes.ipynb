# McCall's Quality Model

```python
import networkx as nx
import matplotlib.pyplot as plt

# Quality attributes and their characteristics from McCall model
quality_attributes = [
    "Correctness",
    "Reliability",
    "Efficiency",
    "Integrity",
    "Usability",
    "Maintainability",
    "Flexibility",
    "Testability",
    "Portability"
]

characteristics = [
    ["Number of defects", "User requirements", "System requirements"],
    ["Mean Time to Failure", "Mean Time to Repair", "Failure rate"],
    ["Resource usage", "Memory usage", "Disk usage"],
    ["Access control", "Data integrity", "Authentication methods"],
    ["User satisfaction", "Learnability", "Ease of use"],
    ["Code readability", "Modularity", "Comment density"],
    ["Customizability", "Configurability", "Extensibility"],
    ["Test coverage", "Test case count", "Test automation"],
    ["Platform independence", "Hardware independence", "Operating system independence"]
]

# Create a directed graph
graph = nx.DiGraph()

# Add nodes for quality attributes
graph.add_nodes_from(quality_attributes, color='lightblue')

# Add nodes for characteristics and connect them to their respective quality attributes
for attr, chars in zip(quality_attributes, characteristics):
    for char in chars:
        graph.add_node(char, color='lightcoral')
        graph.add_edge(attr, char)

# Plot the graph with different colors
colors = [graph.nodes[n]['color'] for n in graph.nodes]

# Increase the figure size to provide more space
plt.figure(figsize=(12, 8))

# Use Kamada-Kawai layout for better spacing
pos = nx.kamada_kawai_layout(graph)

# Plot the graph with different colors
nx.draw(graph, pos, with_labels=True, font_weight='bold', node_size=2000, node_color=colors, font_size=8, font_color='black', edge_color='gray', linewidths=1, arrowsize=20)

# Save the plot as a PNG file
plt.savefig('knowledgegraph_McCall.png')

# Show the plot
plt.show()


```
